```python
import pandas as pd
import os

# Ruta base donde tengas tus CSVs
base_path = "../data/csv/raw/"

# Lista de archivos que me pasaste
csv_files = [
    "ucl_clubs_key_stats_1992_2025.csv",
    "ucl_clubs_goals_stats_1992_2025.csv",
    "ucl_clubs_attacking_stats_1992_2025.csv",
    "ucl_clubs_defending_stats_1992_2025.csv",
    "ucl_clubs_attempts_stats_1992_2025.csv",
    "ucl_clubs_distribution_stats_1992_2025.csv",
    "ucl_clubs_goalkeeping_stats_1992_2025.csv",
    "ucl_clubs_disciplinary_stats_1992_2025.csv"
]

def explore_csv(path):
    print("="*80)
    print(f"📁 Archivo: {path}")
    print("="*80)

    df = pd.read_csv(path)

    # Filas y columnas
    print(f"🔢 Forma: {df.shape} (filas, columnas)\n")

    # Primeras filas
    print("📌 Primeras 5 filas:")
    print(df.head(), "\n")

    # Ultimas filas
    print("📌 Últimas 5 filas:")
    print(df.tail(), "\n")

    # Tipos de datos
    print("🧬 Tipos de columnas:")
    print(df.dtypes, "\n")

    # Columnas
    print("📋 Columnas:")
    print(list(df.columns), "\n")

    # Valores faltantes
    print("❓ Valores nulos:")
    print(df.isnull().sum(), "\n")

    # Estadísticas básicas (solo numéricas)
    print("📊 Descripción numérica:")
    print(df.describe(), "\n")

    return df


# Explorar todos los CSVs
dataframes = {}
for fname in csv_files:
    full_path = os.path.join(base_path, fname)
    if os.path.exists(full_path):
        dataframes[fname] = explore_csv(full_path)
    else:
        print(f"⚠️ No encontrado: {full_path}")
```
